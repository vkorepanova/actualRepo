```python
#%%timeit -n 100
import matplotlib.pyplot as plt

import time
v = []
number = []
import random as rand
from random import randint
start_time = time.time()
for start in range(100):
    def qs(a, left, right):
        if left>=right:
            return a
        i, j = left, right
        x = a[randint(left, right)]
        while i<=j:
            while a[i]<x:
                i+=1
            while a[j]>x:
                j-=1
            if i<=j:
                a[i], a[j] = a[j], a[i]
                i+=1
                j-=1
        qs(a, left, j)
        qs(a, i, right)
        return(a)
    n = randint(100, 1000)
    a = [rand.randint(0,1000) for i in range(n)]
    print(qs(a, 0, len(a)-1))
    v.append(float("%s" % (time.time() - start_time)))
    number.append(n)
plt.xlabel('Кол-во элементов')
plt.ylabel('Время')
plt.title('Временная сложность алгоритма', color = "blue")
plt.plot(v)
plt.ylabel(number)
plt.yscale('log')
plt.show()
```
